```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
student_monthly = pd.read_csv("../data/processed/studentdata.csv")
student_monthly.pop("Unnamed: 0")
```

```python
km = KModes(n_clusters=2, init='Huang', n_init=5, verbose=1)
clusters = km.fit_predict(student_monthly)
student_monthly['Cluster2'] = clusters
```

```python
km = KModes(n_clusters=2, init='random', n_init=5, verbose=1)
clusters = km.fit_predict(student_monthly)
student_monthly['Cluster2'] = clusters

centroids = km.cluster_centroids_
print("Cluster Centroids:")
print(centroids)

for cluster in range(km.n_clusters):
    print(f"\nCluster {cluster}:")
    cluster_data = student_monthly[student_monthly['Cluster2'] == cluster]
    print(cluster_data.describe(include='all'))
```

```python
# Cluster centroids
centroids = km.cluster_centroids_
print("Cluster Centroids:")
print(centroids)
```

```python
# Cluster analysis
for cluster in range(km.n_clusters):
    print(f"\nCluster {cluster}:")
    cluster_data = student_monthly[student_monthly['Cluster2'] == cluster]
    print(cluster_data.describe(include='all'))
```

```python
km4 = KModes(n_clusters=4, init='Huang', n_init=5, verbose=1)
clusters4 = km4.fit_predict(student_monthly)
student_monthly['Cluster4'] = clusters4

centroids4 = km4.cluster_centroids_
print("Cluster Centroids:")
print(centroids4)
```

```python
# Cluster analysis
for cluster in range(km4.n_clusters):
    print(f"\nCluster {cluster}:")
    cluster_data4 = student_monthly[student_monthly['Cluster4'] == cluster]
    print(cluster_data4.describe(include='all'))
```

```python
cost = []
K = range(1,6)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 5, verbose=1)
    kmode.fit_predict(student_monthly)
    cost.append(kmode.cost_)
    
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```
